```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
driver = webdriver.Chrome('/Users/jubong/Downloads/chromedriver')
driver.implicitly_wait(3)


def fread(path, mode='r'):
    f = open(path, mode)
    return f
def fwrite(path, mode='a'):
    f = open(path, mode)
    return f
def fclose(f):
    return f.close()

def similarity(text, p):
    text = text.split(' ')
    text.append("journal")
    cnt = 0
    for ti in text:
        if ti in p:
            cnt += 1
    return cnt
read_f = fread('/Users/jubong/Downloads/journal_주봉이형.txt')
write_f = fwrite('/Users/jubong/Downloads/journal_complete.txt')
while True:
    read_line = read_f.readline()
    if read_line == '':
        break

    splited_read_line = read_line.split('title=')

    r = splited_read_line[1][1:-2]
    rsplited = r.split(' ')
    rsplited.append('issn')
    rstr = ''
    for rs in rsplited:
        rstr += rs
        if not rs == rsplited[-1]:
            rstr += '+'
    driver.get('https://www.google.com/search?q={0}&oq={0}&aqs=chrome..69i57.3118j0j4&sourceid=chrome&ie=UTF-8'.format(rstr))
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    notices = soup.select('div.rc')


    for n in notices:
        url = str(n.a).split('"')[1]
        pattern = re.compile(r'(\d{4})-(\d{4})')
        match = pattern.search(n.text)
        if match is not None:
            issn_number = str(match.group())
        else:
            continue
        issn_number = issn_number.strip()
        read_line = read_line.replace("ISSN=''", "ISSN='"+issn_number+"'")
        print(issn_number)
        if len(issn_number) != 0 and ('wikipedia' not in url):
            url = url.strip()
            read_line = read_line.replace("HOMEPAGE=''", "HOMEPAGE='"+url+"'")
            driver_site = driver.get(url)
            driver.implicitly_wait(3)

            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            title = soup.select('title')
            try:
                publisher = (title[0].text.split('- ')[-1].strip())
                read_line = read_line.replace("PUBLISHER=''", "PUBLISHER='"+publisher+"'")
            except IndexError:
                pass
            notices = soup.select('p')

            ps = list()
            count = list()
            for n in notices:
                #print(n.text)
                #print(similarity(r, str(n.text)))
                ps.append(n.text)
                count.append(similarity(r, str(n.text)))
                #print()
            #print(ps[count.index(max(count))])
            try:
                description = ps[count.index(max(count))].strip()
                read_line = read_line.replace("DESCRIPTION=''", "DESCRIPTION='"+description+"'")
            except ValueError:
                pass
            break
    write_f.write(read_line + '\n')        
read_f.close()
write_f.close()
```
