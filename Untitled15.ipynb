```python
import threading
from scapy.all import *
```

```python
class sniffing(threading.Thread):
    def __init__(self):
        threading.Thread.__init__(self)
        self.sniffed = scapy.plist.PacketList
    def run(self):
        self.sniffed = sniff(timeout=5,filter="tcp", prn=lambda x:x.summary())
```

```python
class ppp(threading.Thread):
    def run(self):
        time.sleep(1)
        for i in range(3):
            print("------------------------------------------------")
```

```python
for i in range(50):
    send = sniffing()
    receive = ppp()
    send.start()
    receive.start()
```
