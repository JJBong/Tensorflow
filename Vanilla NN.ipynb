## 1.Vanilla NN(Single Layer + No Batch Input)

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
print type(mnist.train.images), mnist.train.images.shape   #tensorflow가 아닌 numpy ndarray임을 확인
print type(mnist.train.labels), mnist.train.labels.shape
```

```python
import tensorflow as tf
x = tf.placeholder(tf.float32, [1, 784])   #Input Data
print "x -", x.get_shape()
```

```python
y = tf.placeholder(tf.float32, [1, 10])  
```

```python
W = tf.Variable(tf.zeros([784, 10]))
b = tf.Variable(tf.zeros([1, 10]))
print "W -", W.get_shape()
print "b -", b.get_shape()
```

```python
u = tf.matmul(x, W) + b
print "u -", u.get_shape()
```

```python
z = tf.nn.softmax(u)
```

```python
#error = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(u, y))   #classification 방법
error = tf.reduce_sum((z - y) * (z - y)) / 2   #regression 방법
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.05).minimize(error)
```

```python
saver = tf.train.Saver()
```

```python
init = tf.initialize_all_variables()
with tf.Session() as sess:
    sess.run(init)
    for i in range(55000):
        for j in range(5):
            _, error_value = sess.run([optimizer, error], feed_dict={x: [mnist.train.images[i]], y: [mnist.train.labels[i]]})
        if i % 10000 == 0:
            print error_value
            #print W.eval()[300:310]
#     print sess.run(W)
    save_path = saver.save(sess, './mnist.pd')
```

```python
x = tf.placeholder(tf.float32, [None, 784])   #Input Data
print "x -", x.get_shape()
y = tf.placeholder(tf.float32, [None, 10])
```

```python
W = tf.Variable(tf.zeros([784, 10]))
b = tf.Variable(tf.zeros([1, 10]))
print "W -", W.get_shape()
print "b -", b.get_shape()
```

```python
u = tf.matmul(x, W) + b
print "u -", u.get_shape()
```

```python
entropy = tf.nn.softmax_cross_entropy_with_logits(u, y)
print entropy.get_shape()
print

error = tf.reduce_mean(entropy)
print error.get_shape()
print

total_loss = tf.train.GradientDescentOptimizer(0.5).minimize(error)
```

```python
saver = tf.train.Saver()
```

```python
import math
init = tf.initialize_all_variables()
batch_size = 100
total_batch = int(math.ceil(mnist.train.num_examples/float(batch_size)))
epoch = 5
with tf.Session() as sess:
    sess.run(init)
    for e in range(epoch):
        for i in range(total_batch):
            batch_images, batch_labels = mnist.train.next_batch(10)
            _, error_value = sess.run([total_loss, error], feed_dict={x: batch_images, y: batch_labels})
            if i % 1000 == 0:
                print error_value
    save_path = saver.save(sess, './mnist.pd')
```

## 3.Vanilla NN(Multi Layer + Batch Input)

```python
training_epochs = 40
learning_rate = 0.001
batch_size = 100
display_step = 1

n_input = 784
n_hidden_1 = 256
n_hidden_2 = 256
n_classes = 10

x = tf.placeholder(tf.float32, [None, n_input])
y = tf.placeholder(tf.float32, [None, n_classes])
```

```python
weights = {
    'W1' : tf.Variable(tf.random_normal([n_input, n_hidden_1])),
    'W2' : tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2])),
    'out' : tf.Variable(tf.random_normal([n_hidden_2, n_classes]))
}
biases = {
    'b1' : tf.Variable(tf.random_normal([n_hidden_1])),
    'b2' : tf.Variable(tf.random_normal([n_hidden_2])),
    'out' : tf.Variable(tf.random_normal([n_classes]))
}
```

```python
def multilayer_perceptron(x, weights, biases):
    #Hidden layer with RELU activation
    u_2 = tf.add(tf.matmul(x, weights['W1']), biases['b1'])
    z_2 = tf.nn.relu(u_2)
    #Hidden layer with RELU actiovation
    u_3 = tf.add(tf.matmul(z_2, weights['W2']), biases['b2'])
    z_3 = tf.nn.relu(u_3)
    #Output layer with linear activation
    u_4 = tf.add(tf.matmul(z_3, weights['out']), biases['out'])
    return u_4
pred = multilayer_perceptron(x, weights, biases)
```

```python
error = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(pred, y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(error)
```

```python
init = tf.initialize_all_variables()
prediction_ground_truth = tf.equal(tf.argmax(pred, 1), tf.argmax(y, 1))
#이퀄을 통해 나온 트루, 폴스에서 잘 안된 애들(폴스)을 볼 수 있는(이미지 형태) 함수를 만들어 마지막에 찍어보기!! 
#ex) 함수(prediction_ground_truth.eval(), n) n개를 볼수있도록
#test데이터의 몇번째인지도 볼 수 있도록!! 그리고 ground_truth(예측된 것)와 같이 출력
#싱글과 멀티 등등 계속 폴스가 나는 것은 왜 안될지..도 알 수 있도록
accuracy = tf.reduce_mean(tf.cast(prediction_ground_truth, tf.float32))


with tf.Session() as sess:
    sess.run(init)
    total_batch = int(mnist.train.num_examples/batch_size)
    print "Total batch: %d" % total_batch
    
    for epoch in range(training_epochs):
        for i in range(total_batch):
            batch_images, batch_labels = mnist.train.next_batch(batch_size)
            sess.run(optimizer, feed_dict={x : batch_images, y : batch_labels})
        print "Epoch %d Finished - Accuracy: %f" % (epoch, accuracy.eval({x: mnist.test.images, y: mnist.test.labels}))
    print("Training % Test Finished!")
```
