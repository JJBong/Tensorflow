```python
def to_8_bit(l):
    k = np.array([int(l[i], 16) for i in range(0, len(list(l)[:33]), 2)], dtype=np.float16)
    m = np.array([int(l[i], 16) for i in range(1, len(list(l)[:33]), 2)], dtype=np.float16)
    if len(k) == len(m):
        for i in range(len(k)):
            m[i] += k[i]*16
        return k,1
    else:
        for i in range(len(m)):
            m[i] += k[i]*16
        m = list(m)
        m.append(k[-1])
        m = np.array(m, dtype=np.float16)
        return m
```
